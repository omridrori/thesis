```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv(r'C:\Users\User\Desktop\thesis\vae\experiments\results2.csv')

# Extract the necessary columns
delta = df['delta']
mig_score = df['mig_score']
vae_eval_accuracy = df['factor_vae_eval_accuracy']

# Plot mig_score as a function of delta
plt.figure(figsize=(10, 5))
plt.plot(delta, mig_score, marker='o', linestyle='-', color='b')
plt.title('MIG Score as a Function of Delta')
plt.xlabel('Delta')
plt.ylabel('MIG Score')
plt.grid(True)
plt.show()

# Plot factor_vae_eval_accuracy as a function of delta
plt.figure(figsize=(10, 5))
plt.plot(delta, vae_eval_accuracy, marker='o', linestyle='-', color='r')
plt.title('Factor VAE Evaluation Accuracy as a Function of Delta')
plt.xlabel('Delta')
plt.ylabel('Factor VAE Evaluation Accuracy')
plt.grid(True)
plt.show()
```
